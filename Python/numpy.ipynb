```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

def f(x1,x2, count):
    return np.linspace(x1,x2, count)
    
x = f(-2*np.pi, 2*np.pi, 30)
y = np.cos(x)
cos = np.cos(x)  
arccos = np.arccos(y)

plt.plot(x,y,marker = "o")
plt.title("График функции y = cos(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(-5,5)
plt.ylim(-5,5)
plt.grid()
plt.show()
```

```python
cos # возвращает число (1)
arccos # возвращает угол (0)
```
