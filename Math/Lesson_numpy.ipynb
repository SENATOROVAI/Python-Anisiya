```python
# график линейной функции
import matplotlib.pyplot as plt # библиотека для диограммы
import numpy as np # библиотека для работы с данными
```

```python
# def f(x):
# return 2*x + b # kx+b

x = np.linspace(-5, 5, 10)
y = 2*x + 1 # f(x)
plt.plot(x,y)
plt.title('График линейной функции')
# plt.xlim(-10,10) 
# plt.ylim(-10,10)
plt.axis([-20,20,-20,20]) #масштабирование
plt.grid() #сетка
plt.show()
#ocь(x,y)
```

```python
help(np.linspace)
len(x)  # count of number
# np.linspace?
```

```python
# x = [round(i) for i in np.linspace(-10,10,10)]
x = np.linspace(-10,10,60)
y=x**2 # график параболы
plt.plot(x,y)
plt.title('График парабалы')
plt.xlabel("Ось x")
plt.ylabel("Ось y")
plt.grid()
plt.show()
```

# столбчатая диаграмма 
### используется для сравнения различных показателей


```python
names = np.array(["SENATOROV", "Vitalik"])
numbers = [3, 5]
plt.bar(names, numbers, color='red')  # столбчатая диаграмма
plt.title('Столбчатая диаграмма')
plt.xlabel("Ось x")
plt.ylabel("Ось y")
plt.show()
```

### Круговая диаграмма

```python
import matplotlib.pyplot as plt

# Круговая диаграмма, в которой секторы будут упорядочены и расположены против часовой стрелки:
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]

fig1, ax1 = plt.subplots() # написать без subplots
ax1.pie(sizes, labels=labels, autopct='%1.1f%%')

plt.show()
```
